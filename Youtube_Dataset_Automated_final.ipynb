# Script to automate extracting video_ids for give year range

```python
def Automated_videoId_Extraction(year):
  count = 0
  records = []
  print('Inside Function')
  while True:
    print(count)
    init_val = len(video_ids)
    next_page_token = None
    while True:
      response  = youtube.search().list(
            part = 'id',
            type = 'video',
            q = query,
            relevanceLanguage = Lang,
            maxResults = 50,
            regionCode = region,
            pageToken = next_page_token,
            publishedAfter = f'{year}-01-01T00:00:00Z',
           publishedBefore = f'{year}-12-31T00:00:00Z'
          ).execute()
      print(response['pageInfo'])
      for item in response['items']:
        if item['id']['videoId'] not in video_ids:
          video_ids.append(item['id']['videoId'])
      next_page_token = response.get('nextPageToken')
      if not next_page_token :
        break
    fin_val = len(video_ids)
    count += 1
    result = fin_val - init_val
    records.append(result)
    print(f'result: {result}')
    print(f"records: {records}")
    if records[0] != 0:
      if ((result < 20)| (result/records[0] < 0.05)):
        print('Work Done!')
        break
    else:
      break
```

# Script to extract all the details of the videos for given year range

```python
def snippet_statistics_content_extraction(video_id_list):
  for i in range(0, len(video_id_list), 50):
    video_ids = video_id_list[i:i+50]
    response_2 = youtube.videos().list(
      part = 'statistics,contentDetails,snippet',
      id = ','.join(video_ids)
  ).execute()

    for item in response_2['items']:
      print(item['snippet']['publishedAt'])
      publishedAt.append(item['snippet']['publishedAt'])
      channelId.append(item['snippet']['channelId'])
      title.append(item['snippet']['title'])
      description.append(item['snippet']['description'])
      thumbnails.append(item['snippet']['thumbnails'])
      channelTitle.append(item['snippet']['channelTitle'])
      liveBroadcastContent.append(item['snippet']['liveBroadcastContent'])
      # publishTime.append(item['snippet']['publishTime'])
      try:
        viewCount.append(item['statistics']['viewCount'])
      except:
        viewCount.append('0')
      try:
        likeCount.append(item['statistics']['likeCount'])
      except:
        likeCount.append('0')
      favoriteCount.append(item['statistics']['favoriteCount'])
      try:
        commentCount.append(item['statistics']['commentCount'])
      except:
       commentCount.append('0')
  
      duration.append(item['contentDetails']['duration'])
      dimension.append(item['contentDetails']['dimension'])
      definition.append(item['contentDetails']['definition'])
      caption.append(item['contentDetails']['caption'])
      licensedContent.append(item['contentDetails']['licensedContent'])
      contentRating.append(item['contentDetails']['contentRating'])
      projection.append(item['contentDetails']['projection'])
```

* Time Conversion function

```python
def duration_conversion(duration):
  duration_seconds = parse_duration(duration)
  minutes, seconds = divmod(duration_seconds.seconds,60)
  hours, minutes = divmod(minutes,60)

  return f'{hours}:{minutes}:{seconds}'
```

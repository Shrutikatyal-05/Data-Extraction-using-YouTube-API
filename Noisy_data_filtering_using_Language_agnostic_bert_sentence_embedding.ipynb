Import Language-Agnostic BERT Sentence Embedding model

```python
from sentence_transformers import SentenceTransformer
```

```python
model=SentenceTransformer("sentence-transformers/LaBSE")
```

Generating Sentence Embeddings using Encode Function

```python
query_sentence ="Madhumakhi Palan,Beekeeping, Bee keeping, Bee-keeping, Apiculture, Honey bee farming, Honeybee farming, Bee Farming, Honey production, Honey harvesting, beekeeper, harvest honey, producing honey, honey business, honey packaging, Honey brand, Honey Export, Start A Bee Farm, Market Your Honey, Beekeeping Business"
query_sentence=(" ").join(query_sentence.split(","))
query_sentence
```

```python
query_embedding=model.encode([query_sentence])
```

```python
query_embedding.shape
```

```python
df['description']=df['description'].astype(str)
```

```python
df1=(df['title'] +" " + df['description']).to_frame()
df1.columns=['title_des']
df1
```

```python
sentence_embedding=model.encode(df1['title_des'].tolist())
```

```python
sentence_embedding.shape
```

Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(query_embedding, sentence_embedding)
```

```python
similarities
```

```python
m,n=similarities.shape[0],similarities.shape[1]
```

```python
similarity_score=similarities.reshape(n,m)
similarity_score
```

```python
df['score']=similarity_score
```

```python
df = df.sort_values(by='score', ascending=False)
```

```python
df
```
